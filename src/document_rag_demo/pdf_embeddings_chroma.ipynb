```python
from langchain.document_loaders import PyPDFDirectoryLoader
from langchain_ollama import OllamaEmbeddings
from langchain.vectorstores import Chroma

persist_directory = "../../data/chroma1/"

embeddings = OllamaEmbeddings(
    model="llama3.2",
    temperature=0
)

loader = PyPDFDirectoryLoader("../../data/pdf/")
docs = loader.load()


chroma_db = Chroma.from_documents(
    documents=docs,
    embedding = embeddings,
    persist_directory=persist_directory,
    collection_name="doc_search_demo"
)
```

```python
documents = chroma_db.similarity_search("Taj Mahal")

for document in documents:
    print(f" Page Content : {document.page_content}, [{document.metadata}]")
```

```python
documents = chroma_db.similarity_search_with_score("Taj Mahal")

for document, score in documents:
    print(f"* [SIM={score:3f}] {document.page_content} [{document.metadata}]")
```
